```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
    r"C:\Users\ANANYA\Downloads\netflix_data_analytics\netflix_titles.csv"
)
```

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Not Rated')
df['duration'] = df['duration'].fillna('Unknown')

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df = df.dropna(subset=['date_added'])

df['year_added'] = df['date_added'].dt.year
```

```python
yearly_content = df['year_added'].value_counts().sort_index()

yearly_content.plot(
    kind='line',
    figsize=(10,5),
    marker='o'
)

plt.title("Netflix Content Added Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.grid(True)

plt.show()
```

```python
top_genres = df['listed_in'].value_counts().head(10)

top_genres.plot(
    kind='barh',
    figsize=(10,6)
)

plt.title("Top 10 Genres on Netflix")
plt.xlabel("Count")
plt.ylabel("Genre")

plt.show()
```

```python
top_directors = df['director'].value_counts().head(10)

top_directors.plot(
    kind='barh',
    figsize=(10,6)
)

plt.title("Top 10 Directors on Netflix")
plt.xlabel("Number of Titles")

plt.show()
```

```python
df['type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Movies vs TV Shows")
plt.ylabel("")

plt.show()
```

```python
from collections import Counter

actors = df['cast'].dropna().str.split(', ').sum()

top_actors = Counter(actors).most_common(10)

actors_df = pd.DataFrame(top_actors, columns=['Actor', 'Count'])

actors_df.plot(
    x='Actor',
    y='Count',
    kind='bar',
    figsize=(10,5)
)

plt.title("Top 10 Actors on Netflix")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)

plt.show()
```

```python
release_trend = df['release_year'].value_counts().sort_index()

release_trend.plot(
    kind='line',
    figsize=(10,5)
)

plt.title("Content Release Trend")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")

plt.show()
```
